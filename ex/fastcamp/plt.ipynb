```python
import matplotlib.pyplot as plt
```

```python
#문제 x가 123, y가 456인 그래프를 그리시오
x = [1,2,3]
y = [4,5,6]
plt.figure()
plt.plot(x,y)
```

```python
#plt.plot(x,y,linewidth =10, color = 'red')
#plt.plot(x,y,marker = 'o', markersize = 10)#데이터 위치 
#plt.plot(x,y, linestyle=':')
plt.plot(x,y, 'o')
plt.title('title', fontsize = 20)#타이틀 만들기
plt.xlabel('x')
plt.ylabel('y')
#plt.grid(True)#뒤에 그리드를 넣어줌
plt.grid(True, axis='x')#aixs설정도 가능
```

```python
plt.plot(x,y)
#plt.xlim([1,2])
#plt.ylim([4,5])#x,y최소,최대값 설정
plt.axis([1,2,4,5])#axis를 사용해서도 가능([x_min,x_max,y_min,y_max])
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 25)#글자 크기 변경
```

```python
plt.plot(x,y)
plt.text(2,5,'test', fontsize = 20)#텍스트 삽입도 가능
```

```python
#범례 추가하기
x1=[1,2,3]
y1=[1,2,3]
x2=[1,2,3]
y2=[1,4,7]
'''
plt.plot(x1,y1, color='blue', label = 'data1')
plt.plot(x2,y2, color = 'red', label ='data2')
plt.legend()#범례 추가

'''
#또는
plt.plot(x1,y1, color='blue')
plt.plot(x2,y2, color = 'red')

plt.legend(('data1', 'data2'), loc='upper right')#범례 위치도 추가
```

```python
#subplot / subplots 를 이용한 여러개의 차트 그리기
x1=[1,2,3]
y1=[1,2,3]
x2=[1,2,3]
y2=[1,100,200]

#subplot을 이용한 해결
#subplot(행의 수, 열의 수, 해당 그래프가 그려질 위치)
plt.subplot(1,2,1)
plt.plot(x1,x2)
plt.title('data1')

plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title('data2')
```

```python
#transpose
x1=[1,2,3]
y1=[1,2,3]
x2=[1,2,3]
y2=[1,100,200]

plt.subplot(2,1,1)
plt.plot(x1,x2)
plt.title('data1')

plt.subplot(2,1,2)
plt.plot(x2,y2)
plt.title('data2')
```

```python
#tight_layout을 이용하면 레이아웃이 자동으로 설정됨
x1=[1,2,3]
y1=[1,2,3]
x2=[1,2,3]
y2=[1,100,200]

plt.subplot(2,1,1)
plt.plot(x1,x2)
plt.title('data1')

plt.subplot(2,1,2)
plt.plot(x2,y2)
plt.title('data2')
plt.tight_layout()
```

```python
#subplots를 이용한 방법
fig,axe1 =plt.subplots(nrows=1, ncols=2)
```

```python
#fig
#axe1
axe1[0].plot(x1,y1,color='blue')
axe1[1].plot(x2,y2,color='red')

fig
```

```python
#문제 data1(x:1,2,3,y:1,2,3)과 data2(x:1,2,3,y:1,50,200)을 그래프로 출력하시오
x1=[1,2,3]
y1=[1,2,3]
x2=[1,2,3]
y2=[1,50,200]

fig, axe1=plt.subplots()

axe2= axe1.twinx()#x축을 공유하고 y축을 따로사용
'''
axe1.plot(x1,y1,color='red',label='data1')
axe2.plot(x2,y2,color='blue',label = 'data2')

axe1.set_xlabel('x', fontsize=15)
axe1.set_ylabel('y1', fontsize=15)
axe2.set_ylabel('y2', fontsize=15)
fig
'''
#2가지 범례 추가하기
chart1 = axe1.plot(x1,y1,color='red',label='data1', linestyle='--')
chart2 = axe2.plot(x2,y2,color='blue',label = 'data2',linestyle ='--')

axe1.set_xlabel('x', fontsize=15)
axe1.set_ylabel('y1', fontsize=15)
axe2.set_ylabel('y2', fontsize=15)

chart = chart1+chart2
axe1.legend(chart, ['data1', 'data2'])
fig
```
